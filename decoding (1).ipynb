## Frequency analysis
Now we can get on to analysing the data itself!

```python
# Create a histogram (dictionary) that counts the frequency of each word

# Adds each word in the list to the dictionary. If the word isn't in the dictionary,
# set its value to 0 then add 1. If word already exists, just add 1.
def histogram(words):
    hist = {}
    for word in words:
        hist[word] = hist.setdefault(word,0) + 1
    
    return hist

# Reverse this histogram to obtain dictionary with frequencies as keys.
# Same concept as previous function.
def reverse(words):    
    hist = histogram(words)
    freq_dict = {}
    for key, value in hist.items():
        freq_dict[value] = freq_dict.setdefault(value,[])
        freq_dict[value].append(key)
        
    
    return freq_dict

moby_dict = reverse(moby_list)

# Get list of top 20 most frequent words


def frequency_list(dict):
    frequencies = []
    for freq in dict.keys():
        frequencies.append(freq)
    
    frequencies.sort(reverse=True)
    
    return frequencies

frequencies = frequency_list(moby_dict)    


def first_n(freq_dict, freq_list, n):
    first_n = []
    i = 0
    while len(first_n) < n:   
        for index in freq_list:
            # Since the entries of the dictionary are lists, 
            # we want to return the individual strings.
            for j in freq_dict[index]:
                if len(first_n) < n:
                    first_n.append((j,index))      

    return first_n
    
        
first_twenty = first_n(moby_dict, frequencies, 20)

for i in range(20):
    print(first_twenty[i])
    
```

```python
def tri_met(message, steps, T, annealing=True, dictionary=True, valid_degree=0, progress=True, poly=False):
    best = message
    valid = set()
    T_0 = T
    for i in range(steps + 1):
        if dictionary:
            dictionary = english
            counter = 0
            
            for word in best.split():
                if word in dictionary:
                    counter += 1
            
            # Valid degree allows you to specify how accurate you want the spell checker to be
            if counter == len(best.split())  - valid_degree and counter > 2:
                valid.add(best)

        a = random.randint(0,26)
        b = random.randint(0,26)
        
        # alphabet is a list of letters defined earlier
                
        maps = {} 
        maps[alphabet[a]] = alphabet[b]
        maps[alphabet[b]] = alphabet[a]

        alternative = test_map(best, maps)

        Sm = tri_score(best)
        Sm_p = tri_score(alternative)

        if Sm_p > Sm:
            best = alternative            
        
        # If alternative scores less than the current best, replace best 
        # with the alternative with given probablility.
        elif T != 0:
            if random.random() < np.exp((Sm_p - Sm)/T):
                best = alternative                
        
        if annealing:
            if not poly:
                T -= T/(steps+1)
            else:
                T = T_0 * (1-(i/steps))**5  
                
        if progress:
            if i % 10000 == 0:
                print('Step {}:'.format(i),best,'\n')
    if progress:
        print('Valid sentences:')
        for sentence in valid:
            print(sentence,'\n')
    return valid
```

The metropolis hastings algorithm converges to some 'optimal' solution every time you run it. This is like minimising a function and finding a local minimum. However, we don't want the local minimium but the GLOBAL minimum, so simulated annealing adds an element of randomness to our algorithm to make it more likely converge to the correct solution.

The downside of annealing is that if there is too much of it, the algorithm will become to chaotic to converge to any reasonable solution at all, so tweaking is required (this can also depend on the problem itself).

```python
# Run as many simulations as you'd like. Stores 'valid' solutions over time. The idea of this method is that you 
# get the benefits of simulated annealing while not having to run the experiment multiple times manually.

valid = set()
for i in range(1): 
    valid_dummy = valid
    valid.update(tri_met(m10,150000,2, progress=False))
    if valid != valid_dummy:
        print('Step {}, valid: {}'.format(i+1,valid))
    if i % 10 == 0:
        print('Step',i+1)
print('Valid sentences:')
for sentence in valid:
    print(sentence,'\n')
```
